```python
dataset_dir = "/content/datasets/9 Facial Expressions you need"

label_dirs = [
    f"{dataset_dir}/train/labels",
    f"{dataset_dir}/valid/labels",
    f"{dataset_dir}/test/labels"
]

for label_dir in label_dirs:
    for file in os.listdir(label_dir):
        if not file.endswith(".txt"):
            continue

        file_path = os.path.join(label_dir, file)
        new_lines = []

        with open(file_path, "r") as f:
            for line in f:
                parts = line.strip().split()
                class_id = int(parts[0])

                # mapping
                if class_id == 7:
                    new_class = 1  # sleepy
                else:
                    new_class = 0  # non_sleepy

                new_lines.append(
                    " ".join([str(new_class)] + parts[1:])
                )

        with open(file_path, "w") as f:
            f.write("\n".join(new_lines))
```

```python
data_config = {
    'train': f"{dataset_dir}/train/images",
    'val':   f"{dataset_dir}/valid/images",
    'test':  f"{dataset_dir}/test/images",

    'nc': 2,
    'names': ['non_sleepy', 'sleepy']
}

config_path = "/content/data_config.yaml"
with open(config_path, "w") as f:
    yaml.dump(data_config, f)
```

```python
# Training
print("üî• Memulai Training...")
model = YOLO('yolov8n.pt')

project_path = "/content/drive/MyDrive/proyek_emosi_yolo_2"

if not os.path.exists("/content/drive"):
    project_path = "/content/proyek_emosi_2"

results = model.train(
    data=config_path,
    epochs=10,
    imgsz=640,
    project=project_path,
    name="model_yolo_v8_2",
    device=0
)
```

```python
last_weight_path = "/content/drive/MyDrive/proyek_emosi_yolo_2/model_yolo_v8_2/weights/last.pt"
model = YOLO(last_weight_path)

print("üî• Melanjutkan Training dari Checkpoint Terakhir...")
results = model.train(
    resume=True,
    device=0
)
```
