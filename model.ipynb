```python
from ultralytics import YOLO

# 1. Path ke model terakhir
last_weight_path = "./model_yolo_v8/weights/last.pt"

# 2. Path ke file konfigurasi data Anda (data_config.yaml lokal)
# Ganti dengan path ke file .yaml Anda (yang menunjuk ke data lokal)
data_config_path = "data_config.yaml" 

# Muat model
model = YOLO(last_weight_path)

# Lanjutkan training dengan menyertakan path data secara eksplisit
# Ini jauh lebih robust saat berpindah environment
results = model.train(
    data=data_config_path,  # <-- TAMBAHAN: Menyebutkan lokasi data
    resume=True,
    device=0
)
```
